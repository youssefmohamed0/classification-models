```python
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
iris = load_iris()
x = iris.data[:,[0,1]]
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2 , random_state = True)
```

```python
def display_boundary(model,x_train,y_train):
    disp = DecisionBoundaryDisplay.from_estimator(
    model, x_train, response_method="predict",
    xlabel=iris.feature_names[0], ylabel=iris.feature_names[1],
    alpha=0.5, cmap=plt.cm.coolwarm
    )
    disp.ax_.scatter(x_train[:, 0], x_train[:, 1], c=y_train, edgecolor='k', cmap=plt.cm.coolwarm)
    plt.show()
```

# Using decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(x_train,y_train)
y_pred = tree.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
display_boundary(tree,x_train,y_train)
```

decision tree classifier is usefull when handeling non linear relationships between featuresa and targets

it is also robust to outliers and fast to train

it is however prone to overfitting can be is computationally expensive when it comes to larger datasets

# Using linear SVM classifier


```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
display_boundary(model,x_train,y_train)
```

svm is highly accurate and is usually used for complex problems

it is also usefull when dealing with non linear relationships and dimensional spaces

it is however memory and computation intensive especially when dealing with larger datasets

# Using naive Baye's classifier

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
display_boundary(model,x_train,y_train)
```
