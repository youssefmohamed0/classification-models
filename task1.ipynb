```python
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
iris = load_iris()
x = iris.data[:,[0,1]]
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2 , random_state = True)
```

```python
def display_boundary(model,x_train,y_train):
    disp = DecisionBoundaryDisplay.from_estimator(
    model, x_train, response_method="predict",
    xlabel=iris.feature_names[0], ylabel=iris.feature_names[1],
    alpha=0.5, cmap=plt.cm.coolwarm
    )
    disp.ax_.scatter(x_train[:, 0], x_train[:, 1], c=y_train, edgecolor='k', cmap=plt.cm.coolwarm)
    plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
display_boundary(knn,x_train,y_train)
```

knn is a simple and easy to understand classification algortith that can hanle npth categorial and numerical data

it can be used for both classification and regression tasks

it is however not suitable for larger datasets as it becomes highlky expensive

its performance decreases as the number of features increases
